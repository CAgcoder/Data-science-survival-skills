```python
from flickrapi import FlickrAPI
from PIL import Image
```

```python
key = '82065772535c66b39f32cce8641a4b48'
secret = '105469b74eb6192f'
```

```python
flickr = FlickrAPI(key, secret, format='etree')
```

```python
keyword1 = 'dog'
dogs = flickr.walk(text=keyword1,
                     tag_mode='all',
                     tags=keyword1,
                     extras='url_c',
                     per_page=500,
                     sort='relevance',
                     )
```

```python
urls1 = []
for i, photo in enumerate(dogs):
    if photo.get('url_c')!= None:
        urls1.append(photo.get('url_c'))
    
    if len(urls1) == 1000:
        break
```

```python
keyword2 = "cat"
cats = flickr.walk(text=keyword2,
                     tag_mode='all',
                     tags=keyword2,
                     extras='url_c',
                     per_page=500,
                     sort='relevance',
                     
                     )
```

```python
urls2 = []
for i, photo in enumerate(cats):
    if photo.get('url_c')!= None:
        urls2.append(photo.get('url_c'))
    if len(urls2) == 1000:
        break
```

```python
keyword3 = "tiger"
tigers = flickr.walk(text=keyword3,
                     tag_mode='all',
                     tags=keyword3,
                     extras='url_c',
                     per_page=500,
                     sort='relevance',
                     
                     )
```
